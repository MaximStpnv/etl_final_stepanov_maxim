### movies

```python
docs_movies = list(db['movies'].find())
movies = pd.json_normalize(docs_movies, sep = '_')
movies = movies[["_id", "title", "year", "released", 
                 "runtime", "genres", "countries", 
                 "languages", "imdb_rating", "imdb_votes", 
                 "tomatoes_viewer_rating", "tomatoes_viewer_numReviews", "num_mflix_comments"]]

movies["year"] = movies["year"].astype(str)
movies["year"] = movies["year"].apply(lambda x: x.replace('è','') if not re.search(r'\d{4}.\d{4}', x) else x.replace('è','-'))
movies['imdb_rating'] = pd.to_numeric(movies['imdb_rating'])
movies['imdb_votes'] = pd.to_numeric(movies['imdb_votes']).astype("Int64")
movies['tomatoes_viewer_numReviews'] = pd.to_numeric(movies['tomatoes_viewer_numReviews']).astype("Int64")
```
